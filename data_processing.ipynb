# Data Processing

```python
from nba_api.stats.static import players, teams
from nba_api.stats.endpoints import playercareerstats, teamyearbyyearstats, playerawards

import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
import time 
```

## NBA_API Players Stats

### player career stats

```python
data = players.get_active_players()
```

```python
df=playercareerstats.PlayerCareerStats(player_id=2544).get_data_frames()[0]
time.sleep(2)
dfT=df.loc[df['SEASON_ID']=='2022-23']

# Process 2022-23 season  
dfT_2022_23 = pd.DataFrame()
for i in range(len(data)):
    time.sleep(0.600)
    df = playercareerstats.PlayerCareerStats(player_id=data[i]['id']).get_data_frames()[0]
    df = df.loc[df['SEASON_ID'] == '2022-23']

    dfT_2022_23 = pd.concat([df, dfT_2022_23], ignore_index=True)
```

```python
dfT_2022_23.to_csv("data_2022_23.csv", index = False)
```

## Team Stats

```python
# processing Team stats (2022-23)
team_data = teams.get_teams()

teams_stats = pd.DataFrame()
# error_teams = []

for i in range(len(team_data)): 
        team = teamyearbyyearstats.TeamYearByYearStats(team_id=team_data[i]['id'])
        team_dat = team.get_data_frames()[0]
        teams_stats = pd.concat([teams_stats, team_dat], axis=0, ignore_index=True)
```

```python
teams_stats
```

```python
teams_stats_2022_24 = teams_stats[teams_stats['YEAR'].str.contains('2022-23|2023-24', na=False) ]
teams_stats_2022_24.to_csv("team_2022_24.csv", index = False)
```
